```python
# Import needed libs
import pandas as pd # data processing
import seaborn as sns # data vis
from dataprep.clean import clean_country #standard names for countries
from functools import reduce #merge several df in one action
```

# Creating Data Set

```python
# Read
df_user=pd.read_csv('/kaggle/input/1-gb-internet-price/worldwide internet users - users.csv')
df_speed=pd.read_csv('/kaggle/input/1-gb-internet-price/worldwide internet speed in 2022  - avg speed.csv')
df_price=pd.read_csv('/kaggle/input/1-gb-internet-price/worldwide internet prices in 2022 - IN 2022.csv')
df_gdp=pd.read_csv('../input/countries/GDP.csv')
df_gdp_pc=pd.read_csv('../input/countries/GDP per capita.csv')
df_happy=pd.read_csv('../input/countries/WHR22.csv')
df_pop=pd.read_csv('../input/countries/population.csv')
df_country=pd.read_csv('../input/undata-country-profiles/country_profile_variables.csv')

# Drop columns in different dfs. I did it based on my opinion
df_country=df_country.drop(columns={'Net Official Development Assist. received (% of GNI)', 
                                    'Pop. using improved sanitation facilities (urban/rural, %)',                     
                                    'Pop. using improved drinking water (urban/rural, %)',
                                    'Energy supply per capita (Gigajoules)',
                                    'Energy production, primary (Petajoules)',
                                    'CO2 emission estimates (million tons/tons per capita)',
                                    'Forested area (% of land area)','Threatened species (number)',
                                    'Threatened species (number)',
                                    'Individuals using the Internet (per 100 inhabitants)',
                                    'Mobile-cellular subscriptions (per 100 inhabitants).1',
                                    'Seats held by women in national parliaments %',
                                    'Education: Government expenditure (% of GDP)',
                                    'Education: Secondary gross enrol. ratio (f/m per 100 pop.)',
                                    'Education: Primary gross enrol. ratio (f/m per 100 pop.)',
                                    'Education: Tertiary gross enrol. ratio (f/m per 100 pop.)',
                                    'Health: Total expenditure (% of GDP)',
                                    'Health: Physicians (per 1000 pop.)',
                                    'Refugees and others of concern to UNHCR (in thousands)',
                                    'International migrant stock (000/% of total pop.)',
                                    'Infant mortality rate (per 1000 live births',
                                    'Life expectancy at birth (females/males, years)',
                                    'Urban population growth rate (average annual %)',
                                    'Balance of payments, current account (million US$)',
                                    'International trade: Balance (million US$)',
                                    'International trade: Imports (million US$)',
                                    'International trade: Exports (million US$)',
                                    'Food production index (2004-2006=100)',
                                    'Agricultural production index (2004-2006=100)',
                                    'Labour force participation (female/male pop. %)',
                                    'GDP per capita (current US$)',
                                    'GDP growth rate (annual %, const. 2005 prices)',
                                    'GDP: Gross domestic product (million current US$)',
                                    'Population density (per km2, 2017)',
                                    'Population in thousands (2017)'})
df_happy=df_happy.drop(columns={'RANK',
                                'Whisker-high',
                                'Whisker-low',
                                'Dystopia (1.83) + residual',
                                'Explained by: GDP per capita',
                                'Explained by: Social support',
                                'Explained by: Healthy life expectancy',
                                'Explained by: Freedom to make life choices',
                                'Explained by: Generosity',
                                'Explained by: Perceptions of corruption'})
df_gdp_pc=df_gdp_pc.drop(columns={'Year',
                                  'Unit'})
df_price=df_price.drop(columns={'Country code',
                                'Continental region'})
df_gdp=df_gdp.drop(columns={'Year', 
                            'Unit'})
df_user=df_user.drop(columns={'Subregion', 
                              'Region', 
                              'Population'})     
df_pop=df_pop.drop(columns={'Unit', 
                            'Year'})

# Raname columns for next step
df_speed = df_speed.rename(columns={'Country':'country'})
df_user = df_user.rename(columns={'Country or area':'country'})
df_price = df_price.rename(columns={'Name':'country'})
df_gdp = df_gdp.rename(columns={'Country/Area':'country'})
df_gdp_pc = df_gdp_pc.rename(columns={'Country/Area':'country'})
df_happy = df_happy.rename(columns={'Country':'country'})
df_pop = df_pop.rename(columns={'Country/Area':'country'})
df_country = df_country.rename(columns={'country':'country'})

# Clean countries names
df_speed=clean_country(df_speed, 'country')
df_user=clean_country(df_user, 'country')
df_price=clean_country(df_price, 'country')
df_gdp=clean_country(df_gdp, 'country')
df_gdp_pc=clean_country(df_gdp_pc, 'country')
df_happy=clean_country(df_happy, 'country')
df_pop=clean_country(df_pop, 'country')
df_country=clean_country(df_country, 'country')

# Drop duplicates. I checked which duplicate to drop: first or last.
df_user=df_user.drop_duplicates(subset = 'country_clean', keep="first")
df_speed=df_speed.drop_duplicates(subset = 'country_clean', keep="first")
df_price=df_price.drop_duplicates(subset = 'country_clean', keep="first")
df_gdp=df_gdp.drop_duplicates(subset = 'country_clean', keep="last")
df_gdp_pc=df_gdp_pc.drop_duplicates(subset = 'country_clean', keep="last")
df_happy=df_happy.drop_duplicates(subset = 'country_clean', keep="first")
df_pop=df_pop.drop_duplicates(subset = 'country_clean', keep="first")
df_country=df_country.drop_duplicates(subset = 'country_clean', keep="last")

# Drop columns with old manes
df_user=df_user.drop(columns=['country'])
df_speed=df_speed.drop(columns=['country'])
df_price=df_price.drop(columns=['country'])
df_gdp=df_gdp.drop(columns=['country'])
df_gdp_pc=df_gdp_pc.drop(columns=['country'])
df_happy=df_happy.drop(columns=['country'])
df_pop=df_pop.drop(columns=['country'])
df_country=df_country.drop(columns=['country'])

# Multiply currency to move from USD to Billion USD
df_gdp['GDP, at current prices - US Dollars'] = df_gdp['GDP, at current prices - US Dollars'].apply(lambda x: x*1e-9)
# Rename column
df_gdp=df_gdp.rename(columns={'GDP, at current prices - US Dollars' : 'gdp_billions'})

# Merge sets
df_m=[df_user, df_speed, df_price, df_gdp, df_gdp_pc, df_happy, df_pop, df_country]
df_m2 = reduce(lambda  left,right: pd.merge(left,right,on=['country_clean'], how='inner'), df_m)
df_m2 = df_m2.rename(columns={'country_clean':'country'})

# Delet NaNs
df_m2=df_m2.dropna()

# Reindex after .dropna()
df_m2.reset_index(drop=True, inplace=True) 

df_m2.info()
```

```python
#check how many 'NaN' values I have in each column
df_m2.isna().sum()
```

```python
# Delet symbols. I checked which symbols I should delete
for col in df_m2.columns:
    df_m2[col] = df_m2[col].astype(str) #Change all types to 'object' for possibility to run next lines
    df_m2[col] = df_m2[col].apply(lambda x: x.replace('$','')) #Delete '$' symbol
    df_m2[col] = df_m2[col].apply(lambda x: x.replace(',',''))  #Delete ',' symbol
    df_m2[col] = df_m2[col].apply(lambda x: x.replace('~',''))  #Delete '~' symbol

# Split columns, add new columns to df and drop old columns
df_m2[['Population age 0-14 years, %', 'Population age distribution 60+ years, %']] = df_m2['Population age distribution (0-14 / 60+ years, %)'].str.split('/', expand=True)
df_m2=df_m2.drop(columns={'Population age distribution (0-14 / 60+ years, %)'})    

# Change all possible column types to 'int/float'
for col in df_m2.columns:    
    df_m2[col] = pd.to_numeric(df_m2[col], errors='ignore')  

# Insert space before capital letter.For example, SouthAfrica -> South Africa 
df_m2['Region'] = df_m2['Region'].str.replace( r"([A-Z])", r" \1")  
df_m2['Region'] = df_m2['Region'].str.lstrip() # delet 'space' before fist capital letter

df_m2.info()
```

```python
# Ramane columns
df_m2=df_m2.rename(columns={'Avg \n(Mbit/s)Ookla': 'speed',
                            'Average price of 1GB (USD)' : 'aver_22',
                            'Cheapest 1GB for 30 days (USD)' : 'min_22',
                            'Most expensive 1GB (USD)': 'max_22',
                            'Average price of 1GB (USD – at start of 2020)' : 'aver_20',
                            'NO. OF Internet Plans ' : 'n_plans',
                            'Average price of 1GB (USD  at the start of 2021)' : 'aver_21',
                            'Internet users' : 'users',
                            'GDP, Per Capita GDP - US Dollars' : 'gdp_pc',
                            'Happiness score' : 'happiness',
                            'Population' : 'pop',
                            'Region': 'region',
                            'Surface area (km2)' : 'surf_area',
                            'Sex ratio (m per 100 f, 2017)' : 'ratio_m_to_f',
                            'Economy: Agriculture (% of GVA)':'econ_agr',
                            'Economy: Industry (% of GVA)':'econ_ind',
                            'Economy: Services and other activity (% of GVA)':'econ_serv',
                            'Employment: Agriculture (% of employed)':'work_agr',
                            'Employment: Industry (% of employed)':'work_ind',
                            'Employment: Services (% of employed)': 'work_serv',
                            'Unemployment (% of labour force)':'no_work',
                            'International trade: Exports (million US$)':'trade_exp',
                            'International trade: Imports (million US$)':'trade_imp',
                            'International trade: Balance (million US$)':'trade_bal',
                            'Urban population (% of total population)':'urban_pop',
                            'Fertility rate, total (live births per woman)':'fertility',
                            'Population age 0-14 years, %':'pop_14-',
                            'Population age distribution 60+ years, %':'pop_60+',
                            'Population growth rate (average annual %)':'pop_grow',
                            'gdp_billions':'gdp',
                            'Mobile-cellular subscriptions (per 100 inhabitants)':'sim_to_pop'})

# Create density pop column
df_m2['density'] = df_m2['pop']/df_m2['surf_area']

# Create sim card column
df_m2['sim_card'] = df_m2['pop']/100*df_m2['sim_to_pop']*1e-6

#Create column % inet users from pop
df_m2['user_to_pop'] = df_m2['users']*100/df_m2['pop']

# Reorder columns
df_m2=df_m2[['country',
             'region',
             'surf_area', 
             'gdp', 'gdp_pc', 
             'happiness', 'pop', 
             'density', 'pop_14-', 
             'pop_60+', 'urban_pop', 
             'pop_grow', 'fertility', 
             'ratio_m_to_f', 
             'users', 
             'user_to_pop', 
             'sim_card', 
             'sim_to_pop', 
             'speed', 
             'n_plans', 
             'aver_22',   
             'min_22', 
             'max_22', 
             'aver_21', 
             'aver_20', 
             'work_agr',
             'work_ind', 
             'work_serv', 
             'no_work', 
             'econ_agr', 
             'econ_ind', 
             'econ_serv']]
```
